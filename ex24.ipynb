```python
import numpy as np
import matplotlib.pyplot as plt
X = [0.0, 1.0, 2.0]
Y = [3.0, 3.5, 5.5]
# y_hat = X.T*w + b
w = 0.0
b =0.0
learning_rate = 0.01
epochs =1_000
y_hat = [0.0, 0.0, 0.0] # 초기화
N = len(X) #3 -> 버퍼사이즈
index = 0
sigma1:float = 0.0
sigma2:float = 0.0
for _ in range(epochs):
    y_hat[index] = X[index] * w + b # y_hat = X*w + b
    for _ in range(N):
        sigma1 += X[index] * (y_hat[index] - Y[index])
        sigma2 += y_hat[index] - Y[index]
    dw = (2/N)*sigma1
    db = (2/N)*sigma2
    w = w -1 * learning_rate * dw
    b = -1 * learning_rate * db
    index = (index + 1) %3
    
```
